```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### read csv

```python
df = pd.read_csv("data/user_data.csv", sep=",",)
print(df.shape)
```

#### 2. 男性と女性の人数を表示

```python
print(df["gender"].value_counts())
```

#### 3. ageのヒストグラム

```python
plt.hist(df["age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

#### 4. 年齢の箱ひげ図

```python
plt.boxplot(df["age"])
plt.ylabel("Age")
plt.show()
```
