```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### read csv

```python
df = pd.read_csv("data/temp_tokyo.csv", sep=",", header=0, index_col=0)
display(df)
```

#### 1.describeを使用して各月の統計量をまとめる。

```python
display(df.describe())
```

#### 2.各月の平均

```python
display(df.mean())
```

#### 3.すべての平均

```python
display(df.mean(axis=1))
```

#### 4.1月の不変分散

```python
print("{:.2f}".format(df["Jan"].var()))
```

#### 5.2000年で最も気温が低い月の温度

```python
print(df.loc[:"2000"].min(axis=1))
```

#### 6. 2005年以降のデータの中で最も気温が高い月は何度？

```python
print(df.loc["2005":].max().max())
```

#### 7. 箱ひげ図

```python
plt.boxplot(df,labels=df.columns)
plt.ylabel("Temperature")
plt.show()
```
