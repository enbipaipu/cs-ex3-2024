```python
df =  pd.read_csv("data/customers.csv", sep=",")
print(df.shape)
print(df.info())
display(df.head())
```

```python
display(df[df.duplicated(keep=False)])
```

```python
df2 = df.drop_duplicates().reset_index(drop=True)
display(df2.shape)
```

```python
print(df2.isna().sum())
```

```python
df2["store"] = df2["store"].fillna("Akabane")
print(df2.isna().sum())
```

```python
df3 = df2.dropna().reset_index(drop=True)
print(df3.shape)
display(df3.head())
```

```python
print(df3["store"].value_counts())
df3["store"] = df3["store"].replace(["hakusan", "Hakusn"], "Hakusan")
print(df3["store"].value_counts())
```

```python
df3.to_csv( "data/customers_cleaned.csv", index=False, encoding="utf-8-sig")
```
