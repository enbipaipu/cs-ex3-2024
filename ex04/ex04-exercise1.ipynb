```python
import pandas as pd
```

#### age:年齢
#### gender:性別
#### occupation:職業
#### zip_code:郵便番号

```python
df = pd.read_csv("data/user_data.csv", sep=",")
print(df.info())
display(df.head())
```

```python
display(df[df.duplicated(keep=False)])
```

```python
df = df.drop_duplicates().reset_index(drop=True)
print(df.shape)
display(df.head())
```

```python
print(df.isna().sum())
```

```python
df["occupation"] = df["occupation"].fillna("unknown")
display(df.head())
```

```python
df = df.dropna().reset_index(drop=True)
display(df.head())
```

```python
df["age"] = df["age"].astype("int")
print(df.info())
```

```python
df["birth_year"] = 2024-df["age"]
display(df.head())
```

```python
df = df.drop(columns="zip_code")
display(df.head())
```

```python
print(df["occupation"].value_counts())
```

```python
df["occupation"] = df["occupation"].replace("eductor", "educator")
df["occupation"] = df["occupation"].replace(["Writer", "write"], "writer")
display(df.head())
```

```python
display(df[(df["age"]<10)|(df["age"]>=70)])
```

```python
date = df[(df["occupation"]=="artist")&(df["gender"]=="M")]
print(date["age"].mean())
```

```python
df = df[~(df["occupation"]=="none")].reset_index(drop=True)
display(df.head())
```

```python
df = df.sort_values(by="age", ascending=True)
display(df.head())
```

```python
print(df["age"].idxmin())
print(df["age"].idxmax())
```

```python
df.to_csv( "data/new_user_data.csv", index=False, encoding="utf-8-sig")
```
