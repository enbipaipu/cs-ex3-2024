```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/temp_tokyo.csv", sep=",", header = 0)
display(df1.head())
print(df1.shape)
print(df1.info())
```

#### Q1:1月のデータを取り出し、表示しよう。

```python
print(df1["Jan"])
```

#### Q2:yearとaprの列を取り出す

```python
display(df1[["Year", "Apr"]])
```

#### Q3:2000年のデータが入っている行を取り出す。

```python
print(df1.loc[0])
```

#### Q4:2000年から2003年までのデータが入っている行を取り出し。

```python
display(df1.iloc[0:4])
```

#### Q5:2002年から2004年までの10月から12月までのデータを取り出す。

```python
display(df1.loc[2:4, "Oct":"Dec"])
```

#### Q6:2010年の1月と2月のデータを取り出す。

```python
print(df1.loc[10, "Jan":"Feb"])
```

#### Q7:年と偶数月のデータが降順になるようにデータを取り出す。

```python
display(df1.iloc[:, 0:13:2].sort_values(by="Year", ascending=False))
```
