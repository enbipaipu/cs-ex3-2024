```python
import pandas as pd
pd.options.display.max_columns = 999
pd.options.display.max_rows=999
```

```python
df = pd.read_csv("severe_cases_daily.csv", sep=",")
print(df.info())
display(df.head())
```

```python
print(df.isna().sum())
```

```python
display(df[df.isna().any(axis=1)].head())
```

```python
new_df = df.fillna(0)
```

```python
display(new_df[new_df.isna().any(axis=1)])
```

```python
display(new_df.head())
display(new_df.tail())
```

```python
df_s = new_df
for s in list(new_df.columns)[1:]:
    df_s[s] = new_df[s].replace(["-", ""], "0")
display(df_s.tail())
```

```python
df_s.to_csv("severe_cases_daily_cleaned.csv", index=False, encoding="utf-8-sig")
```
