```python
import pandas as pd
```

```python
df = pd.read_csv("../ex04/new_user_data.csv", sep=",")
print(df.shape)
display(df.head())
```

### 2.男性女性の人数を表示

```python
print(df["gender"].value_counts())
```

### ３．各職業の平均年齢を

```python
df_ = df[["occupation", "age"]]
display(df_.groupby("occupation").mean())
```

### 4.各職業の男女の合計数を表示。

```python
display(pd.crosstab(df["occupation"], df["gender"]))
```

### 5.職業と性別ごとに最高年齢を表示

```python
display(df.pivot_table(index="occupation", columns = "gender", values="age", aggfunc="max"))
```
