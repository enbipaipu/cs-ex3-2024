```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = 999
pd.options.display.max_rows=999
```

```python
df = pd.read_csv("newly_confirmed_cases_daily.csv", sep="," )
print(df.shape)
print(df.info())
display(df.head())
```

### 2.

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df.info())
```

### 3.

```python
plt.plot(df["Date"], df["Hokkaido"], label="Hokkaido")
plt.plot(df["Date"], df["Tokyo"], label="Tokyo" )
plt.xlabel("Date")
plt.ylabel("number")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### 4.

```python
df = df.set_index("Date")
display(df.head())
```

### 5.

```python
df_rolling7d = df.rolling("7D").mean()
display(df_rolling7d.head())
```

### 6.

```python
plt.figure(figsize=(10, 4))
start_date = pd.to_datetime("2021-05-01")
end_date = pd.to_datetime("2023-05-01")

plt.bar(df.index, df["ALL"], label="num")
plt.plot(df_rolling7d.index, df_rolling7d["ALL"],"--r" , label="7d")
plt.xlim(start_date, end_date)
plt.xlabel("Date", fontsize=14)
plt.ylabel("number", fontsize=14)
plt.xticks(rotation=90)
plt.tick_params(labelsize = 14)
plt.legend()
plt.show()
```

### 7

```python
df_w_sum = df.resample("W").mean()
display(df_w_sum.head())
```

### 8

```python
display(df_w_sum[df_w_sum["ALL"]==df_w_sum["ALL"].max()])
```

### 9.

```python
df["day_of_week"] = df.index.dayofweek
display(df.head())
```

### 10

```python
df_2022 = df[df.index.year==2022]
display(df_2022.head())
```

### 11

```python
df_2022_new = df_2022[["Tokyo","day_of_week"]].groupby("day_of_week").sum()
plt.bar(df_2022_new.index, df_2022_new["Tokyo"])
plt.xlabel("Date", fontsize=14)

plt.ylabel("No. ", fontsize=14)
plt.tick_params(labelsize = 14)

plt.show()
```
